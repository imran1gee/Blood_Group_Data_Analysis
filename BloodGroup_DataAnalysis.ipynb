```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Blood group and probability distribution data
blood_groups = ['O-', 'O+', 'A-', 'A+', 'B-', 'B+', 'AB-', 'AB+']

# Donating probabilities for each blood group (no population weights)
donating_probs = [1, 1/2, 1/2, 1/4, 1/2, 1/4, 1/4, 1/8]

# Adjusted receiving probabilities for each blood group
receiving_probs = [1/8, 1/4, 1/4, 1/2, 1/4, 1/2, 1/4, 1]  # AB+ can receive from all groups

# Plotting the histograms for probability distribution (without population weight)

# Create subplots: one row, two columns for the two histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for Donating Probabilities (without population weight)
axes[0].bar(blood_groups, donating_probs, color='blue', alpha=0.7)  # Bar chart with blue color
axes[0].set_title('Donating Probabilities (No Population Weight)')  # Title for the first subplot
axes[0].set_xlabel('Blood Group')  # Label for x-axis
axes[0].set_ylabel('Probability')  # Label for y-axis
axes[0].set_ylim(0, 1)  # Y-axis range from 0 to 1

# Histogram for Receiving Probabilities (without population weight)
axes[1].bar(blood_groups, receiving_probs, color='red', alpha=0.7)  # Bar chart with red color
axes[1].set_title('Receiving Probabilities (No Population Weight)')  # Title for the second subplot
axes[1].set_xlabel('Blood Group')  # Label for x-axis
axes[1].set_ylabel('Probability')  # Label for y-axis
axes[1].set_ylim(0, 1)  # Y-axis range from 0 to 1

# Display both histograms side by side for comparison
plt.tight_layout()  # Adjust the spacing to prevent overlap
plt.show()  # Show the plot
```

```python
# Blood group and world population distribution data
blood_groups = ['O-', 'O+', 'A-', 'A+', 'B-', 'B+', 'AB-', 'AB+']
population_percent = [7, 38, 6, 34, 2, 9, 1, 3]  # World population distribution in percentages

# Calculate the population fractions (weights) by converting percentages to fractions
population_fraction = [p / 100 for p in population_percent]

# Create a DataFrame for the population distribution data
population_data = {
    'Blood Group': blood_groups,
    'Population (%)': population_percent,
    'Population Weight (Fraction)': population_fraction
}

# Convert the population data into a pandas DataFrame for easy display and manipulation
df_population = pd.DataFrame(population_data)

# Display the table of population weights (fractions)
print("Blood Group Population Weight Table:")
print(df_population)

# Plotting the histogram for population weights (fractions)
plt.figure(figsize=(10, 6))  # Set the size of the plot
plt.bar(blood_groups, population_fraction, color='green', alpha=0.7)  # Create a bar chart with green color
plt.title('Blood Group Population Weights')  # Title of the plot
plt.xlabel('Blood Group')  # Label for x-axis
plt.ylabel('Population Weight')  # Label for y-axis
plt.ylim(0, max(population_fraction) * 1.1)  # Set the y-axis range to display the full data

# Display the histogram for population weights
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Show the plot
```

```python
# Blood group and world population distribution data
blood_groups = ['O-', 'O+', 'A-', 'A+', 'B-', 'B+', 'AB-', 'AB+']
population_percent = [7, 38, 6, 34, 2, 9, 1, 3]  # World population distribution in %
population_fraction = [p / 100 for p in population_percent]

# Donating probabilities for each blood group
donating_probs = [1, 1/2, 1/2, 1/4, 1/2, 1/4, 1/4, 1/8]

# Adjusted receiving probabilities for each blood group (reflecting AB+ being universal recipient)
receiving_probs = [1/8, 1/4, 1/4, 1/2, 1/4, 1/2, 1/4, 1]  # AB+ can receive from all groups

# Calculate weighted probabilities for donating (correct)
weighted_donating_probs = [donating_probs[i] * population_fraction[i] for i in range(len(blood_groups))]

# Calculate weighted probabilities for receiving (correct)
weighted_receiving_probs = [receiving_probs[i] * population_fraction[i] for i in range(len(blood_groups))]

# Create a DataFrame for better visualization
data = {
    'Blood Group': blood_groups,
    'Donating Probability (Fraction)': donating_probs,
    'Receiving Probability (Fraction)': receiving_probs,
    'Weighted Donating Probability': weighted_donating_probs,
    'Weighted Receiving Probability': weighted_receiving_probs
}

df = pd.DataFrame(data)

# Display the table
print("Blood Group Probability Distribution Table:")
print(df)

# Plotting the histograms separately

# Create subplots: one row, two columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for Donating Probabilities (weighted)
axes[0].bar(blood_groups, weighted_donating_probs, color='blue', alpha=0.7)
axes[0].set_title('Weighted Donating Probabilities')
axes[0].set_xlabel('Blood Group')
axes[0].set_ylabel('Probability')
axes[0].set_ylim(0, max(weighted_donating_probs) * 1.1)

# Histogram for Receiving Probabilities (weighted)
axes[1].bar(blood_groups, weighted_receiving_probs, color='red', alpha=0.7)
axes[1].set_title('Weighted Receiving Probabilities')
axes[1].set_xlabel('Blood Group')
axes[1].set_ylabel('Probability')
axes[1].set_ylim(0, max(weighted_receiving_probs) * 1.1)

# Display both histograms side by side
plt.tight_layout()
plt.show()

# Side-by-side histogram in a single plot
plt.figure(figsize=(10, 6))
plt.bar(blood_groups, weighted_donating_probs, width=0.4, label='Weighted Donating Probability', color='blue', alpha=0.7, align='center')
plt.bar(blood_groups, weighted_receiving_probs, width=0.4, label='Weighted Receiving Probability', color='red', alpha=0.7, align='edge')

# Adding labels and title
plt.xlabel('Blood Group')
plt.ylabel('Probability')
plt.title('Comparison of Weighted Donating and Receiving Probabilities')
plt.legend()

# Display the side-by-side histogram
plt.tight_layout()
plt.show()
```
